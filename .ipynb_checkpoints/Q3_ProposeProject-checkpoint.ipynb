## Project Propose: Heart Health Level Monitoring

###### Introduction

Providing personalized services to consumers is a significant goal of almost all businesses. As the country focuses on promoting population health and preventing costly health problems, the ability to provide personalized advice for active life style and health monitoring for early disease detection is becoming ever more important for businesses at large and the healthcare industry in particular. In this project we propose to investigate the feasibility of understanding people’s health based on personal activity data captured from wireless and wearable devices. Wristband health/fitness monitors and health apps on the smartphones has gained lots of popularity in recent years. The low-cost and easy-to-wear features make this kind of product easy to track daily activities, record stats data and help users to reach a healthier lifestyle. In contrast to existing analytics approaches that focus on tracking how much exercise a user have done as well as how many calories have burned, we are interested in taking advantage of its heart rate monitoring function captured during various activities to decide user’s heart health condition. The level of heart health is an important indicator of human’s overall health, and also an indicator of a person’s exercise performance in certain conditions. For healthy people or athletics, based on the fluctuations of heart health level, we can advice him/her to adjust exercise routines for reaching overall fitness or specific competition goals more efficiently and safely. And for older people or patients, we can help monitor overall health condition and provide early evidence of cardiovascular diseases.

###### Data

The main resource is the open source PAMAP 2 dataset on the UCI Machine Learning Repository website: (https://archive.ics.uci.edu/ml/datasets/PAMAP2+Physical+Activity+Monitoring).

The dataset contains 9 subjects' activity performing data from 18 activities.

###### Analysis

```python
%matplotlib inline

from pandas import *
import pandas as pd
import pylab
import os.path
import glob
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram

sub =pd.DataFrame()
data = pd.DataFrame()
df = pd.DataFrame()

# Load and preprocess data, calculate HR_activity data as new feature matrix

for i in range(1,10):
    #location = 'C:\Users\mzheng1\Dropbox\z_proposal_shared\DataIncubatorChallenge\Q3\Protocol'
    location = '/Users/mingna/Dropbox/z_proposal_shared/DataIncubatorChallenge/Q3/Protocol'
    dir_path = os.path.join(location, 'subject10'+str(i)+'.dat')
    
    columns = [1,2]
    data = read_table(dir_path,delim_whitespace=True, header = None, usecols=columns)
    # Delete missing HR data
    data.columns = ['activityID','HR']
    d2 = data.dropna(axis=0)
    # Discard tgeg rows labled with activityID: 0 (other: transient activities)
    d3 = d2[d2.activityID !=0]
    # Calculate means and standard deviations of each activity's HR for each subject as the features 
    sub_std = d3.groupby('activityID').std()
    sub_mean = d3.groupby('activityID').mean()
    # Build the new table for HR_Activity data
    sub = concat([sub_mean,sub_std], axis =0)

    # Convert index(activityID) to a column
    sub.reset_index(level=0, inplace=True)
    # Append HR data to the new table
    df = df.append(sub.HR)
# Store the data to data-array 'data', and preprocessing for following clustering
data = df
data.reset_index(drop = True)
# Fill the missing data with the average of the column(all the experiment's subjects)
data = data.fillna(data.mean())
data.reset_index(drop = True)
print (data)
```

```python
# Implement PCA analysis on dataset, see if our data features can seperate subjects into groups.

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X = pca.fit_transform(data)
print pca.explained_variance_ratio_
```

Plot the two components with highest variance ratio [ 0.69161349  0.14392238]. 
From the plot, we can see the data can be separated into different patterns.

```python
plt.scatter(X[:,0], X[:,1])
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('PCA projection')
```

Hierarchical clustering is used in order to build a hierarchy of clusters, and see if different hear rate patterns exist among the subjects. The difference of heart rate between subjects for each activity is calculated as the distance matrix. 

```python
data_dist = pdist(data) # computing the distance
data_link = linkage(data_dist,method='complete') # computing the linkage
dendrogram(data_link)
plt.xlabel('Subjects')
plt.ylabel('Distance')
plt.title('Subjects Hierarchical Clustering');
```

Hierarchical clustering result. 9 experiment’s subjects can be very well divided into 2 categories. This shows using people’s physical activity data to categorize levels of heart health is viable.
