```python
%matplotlib inline

from pandas import *
import pandas as pd
import pylab
import os.path
import glob
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram

sub =pd.DataFrame()
data = pd.DataFrame()
df = pd.DataFrame()

# Load the data, select and calculate HR_activity data

for i in range(1,10):
    #location = 'C:\Users\mzheng1\Dropbox\z_proposal_shared\DataIncubatorChallenge\Q3\Protocol'
    location = '/Users/mingna/Dropbox/z_proposal_shared/DataIncubatorChallenge/Q3/Protocol'
    dir_path = os.path.join(location, 'subject10'+str(i)+'.dat')
    
    columns = [1,2]
    data = read_table(dir_path,delim_whitespace=True, header = None, usecols=columns)
    # Delete missing HR data
    data.columns = ['activityID','HR']
    d2 = data.dropna(axis=0)
    # Discard tgeg rows labled with activityID: 0 (other: transient activities)
    d3 = d2[d2.activityID !=0]
    # Calculate means and standard deviations of each activity's HR for each subject as the features 
    sub_std = d3.groupby('activityID').std()
    sub_mean = d3.groupby('activityID').mean()
    # Build the new table for HR_Activity data
    sub = concat([sub_mean,sub_std], axis =0)

    # Convert index(activityID) to a column
    sub.reset_index(level=0, inplace=True)
    # Append HR data to the new table
    df = df.append(sub.HR)
# Store the data to data-array 'data', and preprocessing for following clustering
data = df
data.reset_index(drop = True)
# Fill the missing data with the average of the column(all the experiment's subjects)
data = data.fillna(data.mean())
data.reset_index(drop = True)
```

```python
# Run Hierarchical Clustering method for clustering, using all the features
data_dist = pdist(data) # computing the distance
data_link = linkage(data_dist,method='complete') # computing the linkage
dendrogram(data_link)
plt.xlabel('Subjects')
plt.ylabel('Distance')
plt.suptitle('Subjects clustering', fontweight='bold', fontsize=14);
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X = pca.fit_transform(data)
print pca.explained_variance_ratio_
```
